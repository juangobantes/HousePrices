Reading the description of the dataset, we can see that a few features (PoolQC, Alley...) have NaN as a possible value which means no pool, no alley...

We need to fill that values somehow so let's find which features have NaN as a value

```python
# columns where NaN values have meaning e.g. no pool etc.
columns_fill = ['PoolQC','MiscFeature','Alley','Fence','MasVnrType','FireplaceQu',
               'GarageQual','GarageCond','GarageFinish','GarageType', 'Electrical',
               'KitchenQual', 'SaleType', 'Functional', 'Exterior2nd', 'Exterior1st',
               'BsmtExposure','BsmtCond','BsmtQual','BsmtFinType1','BsmtFinType2',
               'MSZoning', 'Utilities']

# replace 'NaN' with 'None' in these columns
for col in columns_fill:
    data_train[col].fillna('None',inplace=True)
    data_test[col].fillna('None',inplace=True)
```

```python
# Number of plots (based on the number of categorical features)
num_feats = len(categorical_feats)

# Determine the number of rows and columns for the subplots
rows = (num_feats + 2) // 3  # Adjust this based on preference
cols = 3  # 3 columns per row (you can adjust this too)

# Set up the figure size (width and height can be adjusted based on your needs)
fig, axes = plt.subplots(rows, cols, figsize=(15, 5 * rows))

# Flatten the axes array in case of multiple rows and columns
axes = axes.flatten()

# Iterate over each categorical feature and plot a countplot
for i, feature in enumerate(categorical_feats):
    sns.countplot(data = data_train, x = feature, ax = axes[i])
    axes[i].set_title(f'Countplot of {feature}', fontsize=14)
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Count')

# Remove any empty subplots if the number of features is not a perfect multiple of the grid
for i in range(num_feats, len(axes)):
    fig.delaxes(axes[i])

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Define the number of subplots based on the number of features
n_feats = len(numerical_feats)
n_cols = 3  # You can change this based on your preference
n_rows = (n_feats + n_cols - 1) // n_cols  # This calculates the number of rows

# Create subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(12, 8))

# Flatten the axes array for easier indexing
axes = axes.flatten()

# Plot each feature in a subplot
for i, feat in enumerate(numerical_feats):
    sns.histplot(data_train[feat], bins=30, kde=True, ax=axes[i])
    axes[i].set_title(f'Histogram of {feat}')

# If there are unused subplots, hide them
for i in range(len(numerical_feats), len(axes)):
    fig.delaxes(axes[i])

# Adjust layout
plt.tight_layout()
plt.show()
```
